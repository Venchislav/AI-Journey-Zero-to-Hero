```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linear_function(x, w=2, b=0):
  return w * x + b
```

```python
data = [linear_function(i) for i in range(10_000)]
```

```python
plt.title('Simple Linear Function')
plt.plot(range(10_000), data);
```

Such data can be described with linear regression model:

```python
plt.title('Linear Regression Example:')
plt.scatter(range(400), scatter_data);
plt.plot(range(400), [linear_function(x) for x in range(400)], c='red');

xlabel = 'X'
ylabel = 'y'

plt.show();
```
