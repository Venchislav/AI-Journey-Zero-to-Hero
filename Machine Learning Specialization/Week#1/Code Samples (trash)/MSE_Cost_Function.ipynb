```python
import matplotlib.pyplot as plt
```

```python
x = range(100)
y = range(100)
```

```python
plt.title('Simple Linear data')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y);
```

```python
def simple_function(w):
  return [w * x_i for x_i in x]
```

```python
def mse(w):
  result = 0
  for i in range(len(x)):
    result += (simple_function(w)[i] - y[i]) ** 2

  return 1 / len(x) * result
```

```python
mse(1)
```

```python
loss_values = []

for i in range(-10, 13):
  loss_values.append(mse(i))
```

```python
plt.plot(range(-10, 13), loss_values);
```

This graph is great, as it shows us that loss minimum is at x = 1 (it's 0 there)

```python
y_hat = simple_function(1)
```

```python
plt.title('Simple Linear data')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y)
plt.plot(x, y_hat, c='red');
```
