```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
img = plt.imread('minipuzik.JPG')
plt.axis('off')
plt.imshow(img);
```

```python
print(img.shape)
print(img.size)
```

```python
w, h, d = img.shape
img_arr = img.reshape(w*h, d)
```

```python
img_arr.max()
```

```python
img_arr = img_arr / 255.0
```

```python
from sklearn.utils import shuffle
```

```python
img_arr_sample = shuffle(img_arr, random_state=1)[:1000]
img_arr_sample.size
```

```python
kmeans = KMeans(n_clusters=10, random_state=1)
kmeans.fit(img_arr_sample)
```

```python
labels = kmeans.predict(img_arr)
```

```python
c = kmeans.cluster_centers_
```

```python
len(labels)
```

```python
def recreate_img(c, labels, w, h, d):
  img = np.zeros((w, h, d))
  index = 0

  for i in range(w):
    for j in range(h):
      img[i][j] = c[labels[index]]
      index += 1

  return (img)
```

```python
plt.imshow(recreate_img(c, labels, w, h, d))
plt.axis('off')
plt.show();
```
