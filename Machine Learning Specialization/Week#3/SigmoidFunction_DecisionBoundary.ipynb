```python
import numpy as np
import matplotlib.pyplot as plt
```

**Sigmoid Function is really useful for Classification problems**
___

Formula basis:

$$g(z) = {1 \over 1 + e^{-z}}$$

But full formula is:

$$f_{w,b}(\vec x) = {1 \over 1 + e^{-(\vec w \cdot \vec x + b)}}$$
___

Function outputs values from 0 to 1
___

We can think of it as probability to get 1. <br>
$$P(y=1|x; w, b)$$

```python
z_values_range = np.linspace(-10, 10)
y_values = np.array([1 / (1 + np.e ** -z_values_range[i]) for i in range(len(z_values_range))])
```

```python
plt.title('Sigmoid Function')
plt.plot(range(-10, 10), [0 for _ in range(-10, 10)], c='black')  # x line
plt.plot([0 for _ in range(0, 2)], [0, 1], c='black')  # y line

plt.scatter(0, 0.5, c='red')  # Decision Boundary
plt.plot(z_values_range, y_values);
```
