```python
def gradient_descent_optimization(function):
  x = random.randint(1, 5)
  y = random.randint(1, 5)

  alpha = 0.0001

  print(x, y)

  gradient = scipy.optimize.approx_fprime([x, y], function)

  while math.floor(function([x, y])) != 0:
    x -= alpha * gradient[0]
    y -= alpha * gradient[1]
    print(x, y)
    gradient = scipy.optimize.approx_fprime([x, y], function)

  return  (x, y)
```

```python
print(gradient_descent_optimization(function))
```
