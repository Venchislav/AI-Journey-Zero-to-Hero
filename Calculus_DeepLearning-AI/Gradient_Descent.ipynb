```python
import numpy as np
import matplotlib.pyplot as plt


def function(x):
    return x ** 2


def derivative(y):
    return 2 * y


values = np.arange(-100, 100)
y_values = np.array(list(map(lambda x: function(x), values)))

cur_pos = (80, function(80))
lr = 0.1

for _ in range(1000):
    new_x = cur_pos[0] - lr * derivative(cur_pos[0])
    new_y = function(new_x)

    cur_pos = (new_x, new_y)
    print(new_x, new_y)
```
