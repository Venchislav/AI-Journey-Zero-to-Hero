```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
a = [random.uniform(0, 20) for _ in range(1000)]
b = [a[i] * 2 + random.uniform(0, 10) for i in range(len(a))]
```

```python
plt.scatter(a, b);
```

```python
def corr_coef(a, b):
  sum_deviations = 0
  sum_devs_a = 0
  sum_devs_b = 0
  assert len(a) == len(b)


  for i in range(len(a)):
    sum_deviations += (a[i] - np.mean(a)) * (b[i] - np.mean(b))
    sum_devs_a += (a[i] - np.mean(a)) ** 2
    sum_devs_b += (b[i] - np.mean(b)) ** 2

  cov = sum_deviations / (len(a) - 1)

  r = cov / ((sum_devs_a / (len(a) - 1) * sum_devs_b / (len(b) - 1)) ** 0.5)

  return r
```

```python
print(corr_coef(a, b))
```
