```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1, 12, size=100)
```

```python
plt.hist(data);
```

```python
bins = np.arange(13)-0.5

hist, edges = np.histogram(data, bins=bins)

y = np.arange(1,hist.max()+1)
x = np.arange(12)
X, Y = np.meshgrid(x, y)

plt.scatter(X, Y, c=Y<=hist, cmap="Greys")

plt.show()
```

```python
def get_median(arr):
	if len(arr) % 2 == 1:
		return arr[(len(arr) // 2)]
	else:
		return (arr[(len(arr) // 2)] + arr[(len(arr) // 2) -1]) / 2
```

```python
get_median([1, 2, 3, 4, 5])
```

```python
get_median([1, 2, 3, 4, 5, 6])
```

```python
a = [1, 5, 2, 7, 1, 9, 3, 8, 5, 9]
a = np.array(a)
mean = a.mean()
```

```python
b = []
for i in range(len(a)):
  b.append((a[i] - mean) ** 2)

print((sum(b) / (len(a)-1)) ** 0.5)
```

```python
a = np.array([2, 4, 5, 8, 9, 10, 14, 16])
b = np.array([6, 12, 15, 24, 27, 30, 42, 48])

mean_a = a.mean()
mean_b = b.mean()
```

```python
c = []
d = []
for i in range(len(a)):
  c.append((a[i] - mean_a) ** 2)
  d.append((b[i] - mean_b) ** 2)

x = (sum(c) / (len(a) - 1))**0.5
y = (sum(d) / (len(b) - 1))**0.5
```
